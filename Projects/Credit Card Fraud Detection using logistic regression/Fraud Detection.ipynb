```python
import numpy as np
import pandas as pd
import warnings
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.model_selection import train_test_split as tt
import matplotlib.pyplot as plt
import seaborn as sns

warnings.filterwarnings("ignore")
```

```python
Read csv file
```

```python
df = pd.read_csv('creditcard.csv')
df
```

```python
# Since huge bias in data( Class=0(28000+), Class=1(492))
# Removing Bias
legit = df[df["Class"]==0]
fraud = df[df["Class"]==1]
b = list(fraud.shape)[0]

legit = legit.sample(n=b)

new_df = legit.append(fraud,ignore_index=True)
```

```python
m = new_df.drop(["Class","Time"],axis=1)
n = new_df["Class"]
# sc = StandardScaler()
# m = sc.fit_transform(m)
# print(m)

m_train,m_test,n_train,n_test = tt(m,n,train_size=0.78,stratify=n,random_state=0)
```

```python
log_reg = LogisticRegression(max_iter=50000,random_state=0)
log_reg.fit(m_train,n_train)

n_train_pred = log_reg.predict(m_train)
n_test_pred = log_reg.predict(m_test)
```

```python
acc_train = accuracy_score(n_train,n_train_pred)
acc_test = accuracy_score(n_test,n_test_pred)
print("Training Data Accuracy:",acc_train)
print("Testing Data Accuracy:",acc_test)
```
