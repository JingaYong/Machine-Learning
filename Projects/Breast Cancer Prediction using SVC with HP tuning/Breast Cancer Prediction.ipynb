```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.decomposition import PCA
from mlxtend.plotting import plot_decision_regions
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.model_selection import train_test_split as tt
from sklearn.datasets import load_breast_cancer
```

```python
d = load_breast_cancer()
df = pd.DataFrame(data=d.data,columns=d.feature_names)
df["target"] = d.target
df
```

```python
# Data Processing
x = df.drop('target',axis=1).values
y = df['target'].values
sc = StandardScaler()
x = sc.fit_transform(x)

x_train,x_test,y_train,y_test = tt(x,y,train_size=0.8,stratify=y,random_state=0)
```

```python
clf = SVC()
clf.fit(x_train,y_train)

y_train_pred = clf.predict(x_train)
y_test_pred = clf.predict(x_test)
```

```python
acc_train = accuracy_score(y_train,y_train_pred)
acc_test = accuracy_score(y_test,y_test_pred)
print("Training Data Accuracy:",acc_train)
print("Testing Data Accuracy:",acc_test)
```

```python
# GridSearchCV
p_grid = {'kernel':["linear","poly","rbf","sigmoid"],
          'C':[1,5,10,20]}
gscv = GridSearchCV(estimator=clf,param_grid=p_grid,cv=5)
gscv.fit(x,y)
# print(gscv.cv_results_)
# print(gscv.best_params_)
print("Highest Accuracy:",gscv.best_score_)
```

```python
# RandomizedSearchCV
p_grid = {'kernel':["linear","poly","rbf","sigmoid"],
          'C':[1,5,10,20]}
rscv = RandomizedSearchCV(estimator=clf,param_distributions=p_grid,n_iter=15,cv=5,random_state=0)
rscv.fit(x,y)
# print(rscv.cv_results_)
# print(rscv.best_params_)
print("Highest Accuracy:",rscv.best_score_)
```
