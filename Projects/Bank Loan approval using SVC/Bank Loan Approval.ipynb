```python
import pandas as pd
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split as tt
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("loan_pred.csv")
df
```

```python
# Drop null values
df.dropna(inplace=True)

# label mapping
df.replace({"Loan_Status": {'N': 0, 'Y': 1}}, inplace=True)

df.replace({"Dependents": {'3+': 4}}, inplace=True)

# Categorical data into numerical data
df.replace({"Married": {'No': 0, 'Yes': 1},
            "Self_Employed": {'No': 0, 'Yes': 1},
            "Gender": {'Male': 1, 'Female': 0},
            "Property_Area": {'Rural': 0, 'Semiurban': 1, 'Urban': 2},
            "Education": {'Graduate': 1, 'Not Graduate': 0}}, inplace=True)
```

```python
m = df.iloc[:,1:-1]
n = df.iloc[:,-1]
# print(m.head())
# print(n.head())
# Stratify - equally divides data of each class in train and test sets
m_train,m_test,n_train,n_test = tt(m,n,train_size=0.78,stratify=n,random_state=0)
```

```python
clf = SVC(kernel="linear")
clf.fit(m_train,n_train)
n_train_pred = clf.predict(m_train)
n_test_pred = clf.predict(m_test)
```

```python
acc = accuracy_score(n_train,n_train_pred)
print("SVM Train Accuracy:",acc)
acc_test = accuracy_score(n_test,n_test_pred)
print("SVM Test Accuracy:",acc_test)
```
