```python
import numpy as np
import pandas as pd
import warnings
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.linear_model import LogisticRegression
# from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split as tt
import matplotlib.pyplot as plt
import seaborn as sns

warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('heart.csv')
df
```

```python
m = df.drop(["target"],axis=1)
n = df["target"]

m_train,m_test,n_train,n_test = tt(m,n,train_size=0.78,stratify=n,random_state=0)
```

```python
log_reg = LogisticRegression(max_iter=10000)
log_reg.fit(m_train,n_train)

n_train_pred = log_reg.predict(m_train)
n_test_pred = log_reg.predict(m_test)
```

```python
acc_train = accuracy_score(n_train,n_train_pred)
acc_test = accuracy_score(n_test,n_test_pred)
print("Training Data Accuracy:",acc_train)
print("Testing Data Accuracy:",acc_test)
```

```python
ip = np.array((43,0,1,118,210,0,1,192,0,0.7,2,0,2))
ip = ip.reshape(1,-1)
pred = log_reg.predict(ip)

if pred==1:
    print("Defective Heart")
else:
    print("Healthy Heart")
```
